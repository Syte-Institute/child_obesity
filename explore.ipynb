```python
import pandas as pd
import numpy as np

from IPython.display import Markdown, display
def printmd(string):
    display(Markdown(string))

pd.set_option('display.max_rows', None)

```

```python
# df = pd.read_csv('data/20190321_Masterfile_Kind&Eltern_SALTO.csv', encoding="ISO-8859-1")
df_data = pd.read_excel('data/20190321_Masterfile_Kind&Eltern_SALTO.xlsx', sheet_name=0)
df_desc = pd.read_excel('data/20190321_Masterfile_Kind&Eltern_SALTO.xlsx', sheet_name=1)
```

```python
display(df_data.shape)
```

```python
df_data = df_data.replace('#NULL!', np.nan)
```

```python
df_data
df_desc

# get all t columns
t_columns = { t: [col for col in df_data.columns if col.startswith(t)] for t in ['t1', 't2', 't3', 't4']}
non_t_columns = [col for col in df_data.columns if not col.startswith(('t1', 't2', 't3', 't4'))]

for key in t_columns.keys():
    print('Count total' ,key, len(t_columns.get(key)))
```

```python
df_data[t_columns.get('t1')].isnull().sum().sort_values(ascending=True).head(20).to_frame().merge(df_desc[['Name', 'Beschriftung']], left_index=True, right_on='Name')
```

```python
# display(df_data[t_columns.get('t1')].notnull().sum().sort_values(ascending=False).head(10))
# display(df_data[t_columns.get('t2')].notnull().sum().sort_values(ascending=False).head(10))
# display(df_data[t_columns.get('t3')].notnull().sum().sort_values(ascending=False).head(10))
# display(df_data[t_columns.get('t4')].notnull().sum().sort_values(ascending=False).head(10))

# display(df_data[t_columns.get('t1')].isnull().sum().sort_values(ascending=True).head(10))
# display(df_data[t_columns.get('t2')].isnull().sum().sort_values(ascending=True).head(10))
# display(df_data[t_columns.get('t3')].isnull().sum().sort_values(ascending=True).head(10))
# display(df_data[t_columns.get('t4')].isnull().sum().sort_values(ascending=True).head(10))

df1 = df_data[t_columns.get('t1')].notnull().sum().sort_values(ascending=False).head(50).to_frame().reset_index().merge(df_desc[['Name', 'Beschriftung']], left_on='index', right_on='Name').drop('Name', axis=1).rename(columns={'index': 'Var_t1', 0: 'Count_t1', 'Beschriftung': 'Beschriftung_t1'})
df2 = df_data[t_columns.get('t2')].notnull().sum().sort_values(ascending=False).head(50).to_frame().reset_index().merge(df_desc[['Name', 'Beschriftung']], left_on='index', right_on='Name').drop('Name', axis=1).rename(columns={'index': 'Var_t2', 0: 'Count_t2', 'Beschriftung': 'Beschriftung_t2'})
df3 = df_data[t_columns.get('t3')].notnull().sum().sort_values(ascending=False).head(50).to_frame().reset_index().merge(df_desc[['Name', 'Beschriftung']], left_on='index', right_on='Name').drop('Name', axis=1).rename(columns={'index': 'Var_t3', 0: 'Count_t3', 'Beschriftung': 'Beschriftung_t3'})
df4 = df_data[t_columns.get('t4')].notnull().sum().sort_values(ascending=False).head(50).to_frame().reset_index().merge(df_desc[['Name', 'Beschriftung']], left_on='index', right_on='Name').drop('Name', axis=1).rename(columns={'index': 'Var_t4', 0: 'Count_t4', 'Beschriftung': 'Beschriftung_t4'})

printmd('### t column not null counts, sorted, first 50')
display(df1.sort_values(['Count_t1', 'Var_t1'], ascending=False))
display(df2.sort_values(['Count_t2', 'Var_t2'], ascending=False))
display(df3.sort_values(['Count_t3', 'Var_t3'], ascending=False))
display(df4.sort_values(['Count_t4', 'Var_t4'], ascending=False))
```
